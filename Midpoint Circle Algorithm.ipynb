```python
import matplotlib.pyplot as plt

def midpoint_circle(radius):
    x, y = 0, radius
    d = 1 - radius
    x_values, y_values = [], []

    def plot_circle_points(x, y):
        x_values.extend([x, -x, x, -x, y, -y, y, -y])
        y_values.extend([y, y, -y, -y, x, x, -x, -x])

    plot_circle_points(x, y)

    while x < y:
        if d < 0:
            d += 2 * x + 3
        else:
            d += 2 * (x - y) + 5
            y -= 1
        x += 1
        plot_circle_points(x, y)

    return x_values, y_values

# Example usage
x_values, y_values = midpoint_circle(10)
plt.scatter(x_values, y_values)
plt.title("Midpoint Circle Algorithm")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
