```python
import matplotlib.pyplot as plt

def midpoint_line(x1, y1, x2, y2):
    x_values, y_values = [], []

    dx = x2 - x1
    dy = y2 - y1
    d = dy - (dx / 2)
    x, y = x1, y1

    x_values.append(x)
    y_values.append(y)

    while x < x2:
        x += 1
        if d < 0:
            d += dy
        else:
            y += 1
            d += (dy - dx)
        x_values.append(x)
        y_values.append(y)

    return x_values, y_values

# Example usage
x_values, y_values = midpoint_line(2, 2, 10, 6)
plt.plot(x_values, y_values, marker="o")
plt.title("Midpoint Line Algorithm")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
